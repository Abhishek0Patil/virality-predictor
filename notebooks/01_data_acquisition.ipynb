```python
import pandas as pd
from pytrends.request import TrendReq
from pytrends import exceptions # <--- 1. IMPORT THE EXCEPTIONS MODULE
import time
import os

# ==============================================================================
# ===== THE 80-TREND DATASET FOR A ROBUST VIRALITY PREDICTOR =================
# ==============================================================================

TRENDS_TO_FETCH = {
    # --- International: Tech & Business (15) ---
    'ChatGPT': ('2022-11-20', '2023-07-01', ''),
    'Metaverse': ('2021-10-25', '2022-06-01', ''),
    'GameStop stock': ('2021-01-15', '2021-04-15', ''),
    'Elon Musk Twitter': ('2022-10-01', '2023-01-31', ''),
    'Sora AI': ('2024-02-15', '2024-05-15', ''),
    'Web3': ('2021-10-01', '2022-09-01', ''),
    'BeReal': ('2022-04-01', '2023-01-31', ''),
    'Threads app': ('2023-07-01', '2023-09-30', ''),
    'Clubhouse app': ('2020-12-01', '2021-09-30', ''),
    'Shein': ('2021-01-01', '2023-01-01', ''),
    'NFT': ('2021-01-01', '2022-12-31', ''),
    'SPAC': ('2020-09-01', '2022-01-31', ''),
    'quiet quitting': ('2022-07-01', '2023-02-28', ''),
    'Apple Vision Pro': ('2023-06-01', '2024-03-31', ''),
    'Bitcoin price': ('2020-10-01', '2022-06-30', ''),

    # --- International: Movies, TV & Culture (19) ---
    'Barbenheimer': ('2023-06-15', '2023-08-30', ''),
    'Taylor Swift Eras Tour': ('2023-03-01', '2023-09-30', ''),
    'Squid Game': ('2021-09-01', '2022-02-28', ''),
    'Wordle': ('2021-12-01', '2022-07-31', ''),
    'Wednesday series': ('2022-11-15', '2023-03-15', ''),
    'Tiger King': ('2020-03-15', '2020-07-31', ''),
    'Among Us': ('2020-07-01', '2021-03-31', ''),
    'Fall Guys': ('2020-08-01', '2021-01-31', ''),
    'Elden Ring': ('2022-02-15', '2022-07-31', ''),
    'fidget spinner': ('2017-03-01', '2017-09-30', ''),
    'Pokemon Go': ('2016-07-01', '2017-01-31', ''),
    'Bridgerton': ('2020-12-20', '2021-04-30', ''),
    'The Mandalorian': ('2019-11-01', '2020-02-29', ''),
    'Parasite movie': ('2019-10-01', '2020-03-31', ''), # Oscar bump
    'Cyberpunk 2077': ('2020-11-15', '2021-04-30', ''),
    'Animal Crossing': ('2020-03-15', '2020-09-30', ''),
    'corn song': ('2022-08-15', '2022-10-31', ''), # Meme
    'ice bucket challenge': ('2014-07-15', '2014-09-30', ''), # Classic viral challenge
    'Harlem Shake': ('2013-02-01', '2013-04-30', ''), # Classic meme
    
    # --- International: Global Events (10) ---
    'Ukraine war': ('2022-02-20', '2022-06-30', ''),
    'Queen Elizabeth death': ('2022-09-01', '2022-10-15', ''),
    'Suez Canal ship': ('2021-03-20', '2021-04-30', ''),
    'Area 51 raid': ('2019-06-15', '2019-10-15', ''),
    'James Webb Telescope': ('2021-12-20', '2022-08-31', ''),
    'Brexit': ('2016-06-01', '2016-08-31', ''),
    'solar eclipse 2017': ('2017-08-01', '2017-09-15', ''),
    'Panama Papers': ('2016-04-01', '2016-05-31', ''),
    'COVID-19 symptoms': ('2020-03-01', '2020-07-31', ''),
    'Black Lives Matter': ('2020-05-25', '2020-09-30', ''),
    
    # --- Indian: Movies & Culture (15) ---
    'Jawan movie': ('2023-08-15', '2023-11-30', 'IN'),
    'Naatu Naatu': ('2023-01-01', '2023-04-15', 'IN'),
    'Pathaan movie': ('2023-01-15', '2023-04-30', 'IN'),
    'KGF Chapter 2': ('2022-04-01', '2022-07-31', 'IN'),
    'Brahmastra': ('2022-08-15', '2022-11-30', 'IN'),
    'Shark Tank India': ('2022-01-01', '2022-03-31', 'IN'),
    'Manike Mage Hithe': ('2021-07-01', '2021-12-31', 'IN'),
    'Pushpa movie': ('2021-12-10', '2022-03-31', 'IN'),
    'Mirzapur series': ('2020-10-15', '2021-01-15', 'IN'),
    'Bigg Boss': ('2022-10-01', '2023-02-28', 'IN'), # Use a specific season
    'RRR movie': ('2022-03-20', '2022-07-31', 'IN'),
    'Scam 1992': ('2020-10-01', '2021-01-31', 'IN'),
    'Pasoori': ('2022-05-01', '2022-10-31', 'IN'),
    'Kacha Badam': ('2022-01-15', '2022-04-30', 'IN'),
    'Dhinchak Pooja': ('2017-05-01', '2017-08-31', 'IN'),
    
    # --- Indian: News & Events (10) ---
    'NEET exam leak': ('2024-05-10', '2024-06-20', 'IN'),
    'Rahul Gandhi disqualification': ('2023-03-23', '2023-05-31', 'IN'),
    'India vs Pakistan World Cup': ('2023-10-01', '2023-11-01', 'IN'),
    'Farmers protest India': ('2020-11-15', '2021-04-30', 'IN'),
    'G20 summit Delhi': ('2023-08-15', '2023-09-30', 'IN'),
    'Demonetisation India': ('2016-11-01', '2017-02-28', 'IN'),
    'Citizenship Amendment Act': ('2019-12-01', '2020-03-31', 'IN'),
    'Chandrayaan-3': ('2023-07-10', '2023-09-30', 'IN'),
    'Paytm crisis': ('2024-01-20', '2024-04-30', 'IN'),
    'Agnipath scheme': ('2022-06-10', '2022-08-15', 'IN'),
}


OUTPUT_DIR = '../data'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Data Fetching with Retry Logic ---
# Initialize pytrends
pytrends = TrendReq(hl='en-US', tz=360)

for keyword, (start_date, end_date, geo) in TRENDS_TO_FETCH.items():
    geo_name = "Worldwide" if geo == "" else geo
    print(f"Fetching Google Trends for: '{keyword}' (Geo: {geo_name})")
    
    timeframe = f'{start_date} {end_date}'
    
    # --- 2. START OF THE RETRY LOOP ---
    attempts = 0
    max_attempts = 5
    backoff_time = 60  # Initial wait time in seconds (1 minute)
    
    while attempts < max_attempts:
        try:
            # The core request logic is now inside the try block
            pytrends.build_payload(kw_list=[keyword], timeframe=timeframe, geo=geo)
            gtrends_df = pytrends.interest_over_time()

            if not gtrends_df.empty and keyword in gtrends_df.columns:
                gtrends_df.rename(columns={keyword: 'gtrends_volume', 'isPartial': 'is_partial'}, inplace=True)
                
                safe_keyword = keyword.replace(" ", "_").replace(":", "")
                geo_suffix = f"_{geo}" if geo else "_WW"
                file_path = os.path.join(OUTPUT_DIR, f"{safe_keyword}{geo_suffix}_gtrends.csv")
                
                gtrends_df.to_csv(file_path)
                print(f"  -> Saved to {file_path}")
            else:
                print(f"  -> Could not fetch or data was empty for {keyword}")

            # If successful, break out of the while loop for this keyword
            break 
            
        except exceptions.TooManyRequestsError:
            # This is our specific error handler
            print(f"  -> Rate limit hit. Waiting for {backoff_time} seconds before retrying...")
            time.sleep(backoff_time)
            attempts += 1
            backoff_time *= 2  # Exponentially increase wait time for next potential failure
            
        except Exception as e:
            # Handle other potential errors (e.g., connection issues)
            print(f"  -> An unexpected error occurred: {e}")
            # For other errors, we might not want to retry, so we break
            break
    
    if attempts == max_attempts:
        print(f"  -> FAILED to fetch data for '{keyword}' after {max_attempts} attempts.")
    # --- END OF THE RETRY LOOP ---

    # We still keep a small, polite sleep between successful requests
    time.sleep(10) # <-- Increased the base sleep time to 10 seconds

print("\nData acquisition complete.")
```
